# Pregunta 13
Ordenar un DataFrame según una o más columnas. Por ejemplo, ordenar empleados por departamento y salario en orden descendente.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F 

spark = SparkSession.builder.appName("Pregunta 13").getOrCreate()
df = spark.read.csv("./datasets/employees.csv", header=True, inferSchema=True)
```

```python
df.show(5)
```

```python
df = df.orderBy(F.col("Department"), F.col("Salary").desc())
df.show(5)
```
