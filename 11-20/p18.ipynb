# Pregunta 18
Dado un DataFrame con Customer Name, Order Time y Price, calcula el total, el promedio y el recuento de transacciones por cliente. Luego filtra aquellos con mÃ¡s de 5 transacciones.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F 
from pyspark.sql import Window

spark = SparkSession.builder.appName("Pregunta 18").getOrCreate()

df = spark.read.csv("./datasets/restaurant_orders copy.csv", header=True, inferSchema=True)
```

```python
df.show(5)
```

```python
df2 = df.groupBy(F.col('Customer Name')).agg(F.sum('Price').alias('monto sumado'),
                                       F.avg('Price').alias('monto promedio'),
                                       F.count('*').alias('conteo'))
df2.filter(F.col('conteo') > 5).show()
```
