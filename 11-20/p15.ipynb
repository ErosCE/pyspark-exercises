# Pregunta 15
Use la función rank() para clasificar a los empleados según el salario dentro del departamento.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F 
from pyspark.sql import Window

spark = SparkSession.builder.appName("Pregunta 13").getOrCreate()

df = spark.read.csv("./datasets/employees.csv", header=True, inferSchema=True)
```

```python
df.show(10)
```

```python
# El rank a los empates les asigna el mismo numero y los siguientes saltan ese numero que no se asigno ejemplo rank 15 y 15 el siguiente es 17
window_spec =  Window.partitionBy(F.col("Department")).orderBy(F.col("Salary").desc())
df = df.withColumn("Rank", F.rank().over(window_spec))
df.show(20)
```
