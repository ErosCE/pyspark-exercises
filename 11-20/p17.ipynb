# Pregunta 17
Su DataFrame tiene demasiadas particiones (por ejemplo, después de leer de una fuente grande). Necesita reducir el número de particiones antes de escribir para optimizar el rendimiento.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F 
from pyspark.sql import Window

spark = SparkSession.builder.appName("Pregunta 17").getOrCreate()

df = spark.read.csv("./datasets/World-Stock-Prices-Dataset.csv", header=True, inferSchema=True)
```

```python
# Ver numero de particiones
print(df.rdd.getNumPartitions())
```

```python
# Para que al momento de escribir el DataFrame tenga 5 particiones y no 11 como antes
df = df.coalesce(5)
```
