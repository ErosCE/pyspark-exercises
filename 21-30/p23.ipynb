# Pregunta 23
Reemplazar el nombre del departamento 'IT' por 'Data Group'.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder.appName('Pregunta 23').getOrCreate()

df = spark.read.csv("./datasets/employees.csv",header=True,inferSchema=True)
```

```python
df.show(5)
```

```python
df_updated = df.withColumn("Department", F.when(F.col("Department") == "IT", "Data Group").otherwise(F.col("Department"))) 
df_updated.show(15)
```
