# Pregunta 21
Filtrar clientes cuyos nombres comienzan con 'A'.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder.appName("Pregunta 21").getOrCreate()
df = spark.read.csv('./datasets/employees.csv', header=True, inferSchema=True)
```

```python
df.show(5)
```

```python
# con like
# df.filter(F.col('Name').like('A%')).show()
df.filter(F.col('Name').startswith('A')).show()
```
