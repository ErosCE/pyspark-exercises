# Pregunta 22
Calcular el porcentaje de contribuci√≥n salarial de cada empleado a su departamento.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql import Window

spark = SparkSession.builder.appName("Pregunta 22").getOrCreate()
df = spark.read.csv('./datasets/employees.csv', header=True, inferSchema=True)
```

```python
df = df.select(['Name','Department','Salary'])
df.show(10)
```

```python
window_spect = Window.partitionBy("Department")
df2 = df.withColumn("Suma",F.sum('Salary').over(window_spect))
df2 = df2.withColumn("Porcentaje del Departamento", (F.col('Salary')/F.col('Suma'))*100)
df2.show()
```
